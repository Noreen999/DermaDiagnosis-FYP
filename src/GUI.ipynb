```python
# ------------------------------
# Gradio GUI for 2-Stage Skin Disease Classifier
# ------------------------------

import gradio as gr
import numpy as np
import matplotlib.pyplot as plt


# ------------------------------
# Function to generate bar plots
# ------------------------------
def plot_probabilities(labels, probs, title):
    fig, ax = plt.subplots(figsize=(6, 3))
    ax.barh(labels, probs * 100)
    ax.set_xlim(0, 100)
    ax.set_xlabel("Confidence (%)")
    ax.set_title(title)
    for i, v in enumerate(probs * 100):
        ax.text(v + 1, i, f"{v:.2f}%", color='black', va='center')
    plt.tight_layout()
    return fig

# ------------------------------
# Function for Gradio
# ------------------------------
def gradio_predict(img_path):
    result = predict_skin_disease(img_path)

    # Main text output
    output_text = f"""
Predicted Group: {result['Predicted Group']} ({result['Group Confidence']*100:.2f}%)
Predicted Disease: {result['Predicted Disease']} ({result['Disease Confidence']*100:.2f}%)
"""

    # Bar plots
    group_probs = result.get("Group_Probs", None)
    disease_probs = result.get("Disease_Probs", None)

    # If your predictor doesn't return full probabilities, we create dummy arrays
    if group_probs is None:
        group_probs = np.array([result['Group Confidence'] if i == 0 else 0 for i in range(5)])
    if disease_probs is None:
        disease_probs = np.array([result['Disease Confidence'] if i == 0 else 0 for i in range(3)])

    group_labels = ["Pox", "Pigmentation", "Inflammatory", "Healthy", "Acneiform"]
    disease_labels = ["Class1", "Class2", "Class3"]  # Replace with actual Stage2 class names if needed

    fig_group = plot_probabilities(group_labels, group_probs, "Group Prediction Confidence")
    fig_disease = plot_probabilities(disease_labels, disease_probs, f"{result['Predicted Group']} Disease Confidence")

    return output_text, fig_group, fig_disease

# ------------------------------
# Gradio Interface
# ------------------------------
demo = gr.Interface(
    fn=gradio_predict,
    inputs=gr.Image(type="filepath"),
    outputs=[gr.Textbox(label="Prediction"), gr.Plot(label="Group Probabilities"), gr.Plot(label="Disease Probabilities")],
    title="DermaDiagnosis: Two-Stage Skin Disease Classifier",
    description="Upload a skin image. The system predicts the disease group and specific disease, with confidence scores and bar charts."
)

# Launch GUI
if __name__ == "__main__":
    demo.launch()
```
